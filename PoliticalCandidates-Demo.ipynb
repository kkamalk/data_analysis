```python
full_url = 'https://github.com/PacktPublishing/Python-Data-Cleaning-Cookbook/blob/master/Chapter02/data/candidatenews.json?raw=true'
response = urllib.request.urlopen(full_url)
candidatenews = json.load(response)
```

```python
#Update the all the entries to "The Hill" for all the entries 
#in which the key is "source" and the value is "TheHill"
for dictEntry in candidatenews:
    dictEntry.update((k, "The Hill") for k, v in dictEntry.items()
    if k == "source" and v == "TheHill")
```
