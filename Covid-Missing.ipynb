# EDA - covidtotals_missing

```python
covidtotals_missing = pd.read_csv('https://github.com/PacktPublishing/Python-Data-Cleaning-Cookbook/raw/master/Chapter04/data/covidtotalswithmissings.csv')
covidtotals_missing
```

```python
# check index candidate
covidtotals_missing['iso_code'].nunique()
print()
covidtotals_missing['iso_code'].count()
```

```python
covidtotals_missing.set_index("iso_code", inplace=True)
covidtotals_missing.head(2)
```

```python
totvars = ['location','total_cases','total_deaths','total_cases_pm','total_deaths_pm']
demovars = ['population','pop_density','median_age','gdp_per_capita', 'hosp_beds']
```

```python
covidtotals_missing[demovars].isnull()
```

```python
# axis=0 is the default value (columns)
missCounts_X = covidtotals_missing[demovars].isnull().sum(axis=0)
missCounts_X
```

```python
# by rows, which tells you the number of missing by country
missCounts_Y = covidtotals_missing[demovars].isnull().sum(axis=1)
missCounts_Y
```

```python
# this gives us a count of countries missing 0, 1, ..4
missCounts_Y.value_counts()
```

```python
# select the countries that have more than 4 values missing 
covidtotals_missing.loc[missCounts_Y>3, :]
```

```python
covidtotals_missing.loc[missCounts_Y>3, demovars]
```

```python
missCounts_X.value_counts()
```

```python
covidtotals_missing[totvars].isnull().sum(axis=0)
```

```python
totvarsMissCounts = covidtotals_missing[totvars].isnull().sum(axis=1)
totvarsMissCounts
```

```python
totvarsMissCounts.value_counts()
```

```python
covidtotals_missing.loc[totvarsMissCounts>0]
```

```python
# If I decide to replace NaN values with calculated value
covidtotals_missing.total_cases_pm.fillna(covidtotals_missing.total_cases / (covidtotals_missing.population/1000000), 
                                          inplace=True)

covidtotals_missing.total_deaths_pm.fillna(covidtotals_missing.total_deaths / (covidtotals_missing.population/1000000), 
                                          inplace=True)
```
