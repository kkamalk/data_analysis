# Interrogate the data

## In which event did Michael Jordan win a medal?

```python
mj.City
```

## In which event did Michael Phelps win a medal?

```python
mp['Event'].value_counts()
```

## Select the male, gold medal winners in boxing and sort by athlete

```python
boxing = olympics[(olympics.Medal == 'Gold') & (olympics.Gender == 'Men') & (olympics.Sport == 'Boxing')]
boxing.sort_values(by='Athlete')
```

```python
boxing.sort_values(by='Athlete', ascending=False)
```

### Which five countries have won the most medals since 1984?

```python
# First, create a DataFrame filtered by year 
rec = olympics[olympics.Edition >= 1984]
rec
```

```python
# Top 5 countries with medal count
rec.NOC.value_counts().head(5)
```

## Countries with the least number of medals

```python
# Bottom 30 countries with medal count
rec.NOC.value_counts().tail(30)
```

### Display the male gold medal winners for the 100m Track & Field sprint event over the years. List the results starting with the most recent. Show the Olympic City, Edition, Athlete and the country they represent.

```python
gmh = olympics[(olympics.Gender == 'Men') & (olympics.Medal == 'Gold') & (olympics.Event == '100m')]
gmh.sort_values('Edition',ascending=False)[['City','Edition','Athlete','NOC']]
```

### Plot the number of medals achieved by the Chinese team in Beijing 2008 

```python
chn = olympics[(olympics.Edition == 2008) & (olympics.NOC == 'CHN')]
chn
```

```python
sns.countplot(data=chn, x='Gender')
```

```python
sns.countplot(data=chn, x='Medal')
```

#### You can choose different color maps
- https://matplotlib.org/_images/colormaps_reference_03.png

```python
sns.countplot(data=chn, x='Gender',palette='bwr')
```

### Plot the number of Gold, Silver and Bronze medals for each gender. 

```python
sns.countplot(x='Medal',data=chn, hue='Gender',palette='coolwarm',order=['Gold','Silver','Bronze']);
```
