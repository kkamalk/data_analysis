```python
# one way to solve it 
m=nls97.wageincome.mean()
s=nls97.wageincome.std()
up=m+(3*s)
down=m-(3*s)
nls97[(nls97.wageincome>=down) &(nls97.wageincome<=up)]
```

```python
# Check high wages that are three standard deviiatioins above the mean
highwages = nls97.loc[nls97.wageincome > (nls97.wageincome.mean() + 3*nls97.wageincome.std())]
highwages
```

***Notice the wageincome values are identical***

```python
# Check for outliers that are three standard deviations above or below the mean
mean = nls97['wageincome'].mean()
std = nls97['wageincome'].std()

outliers = nls97[(nls97['wageincome'].abs() - mean).abs() > 3*std]
outliers
```
